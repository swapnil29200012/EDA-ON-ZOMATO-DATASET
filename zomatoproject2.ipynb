```python
import pandas as pd
import numpy as np

import matplotlib.pyplot  as plt
%matplotlib inline
```

```python
import seaborn as sns 
```

```python
df = pd.read_csv('zomato.csv', encoding = 'latin-1')
```

```python
df.describe()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df_countrycode=pd.read_excel('Country-Code.xlsx')
```

```python
df_countrycode
```

```python
fdf=pd.merge(df,df_countrycode, on ='Country Code', how = 'left')
```

```python
fdf.info()
```

```python
country_name =fdf.Country.value_counts().index
```

```python
country_value=fdf.Country.value_counts().values
```

```python
#piechart of top 3 countries
plt.pie(country_value[:3], labels = country_name[:3])
```

```python
#OBSERVATION = Zomato has maximum number of buisness in INDIA followed by US and then UK
```

```python
fdf.info()
```

```python
ratings =fdf.groupby(['Aggregate rating', 'Rating color' ,'Rating text']).size().reset_index().rename(columns={0:'rating_counts'})
```

```python
ratings
```

```python
#observation
#when rating is between 4.5-4.9 , food is excellent
#when rating is between 4.0-4.4 , food is v.good
#when rating is between 3.5-3.9 , food is good
#when rating is between 2.5-3.4 , food is average
#when rating is between 1.8-2.4 , food is poor
```

```python

sns.barplot(x="Aggregate rating",data=ratings,y="rating_counts", hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

```python
#from this we understand that 
#1-majority of the people have not rated the food.
#2-then the next set of grou0p have  rated the food average 
```

```python
sns.countplot(x='Rating color',hue='Rating color' ,data= ratings,palette=['blue','red','orange','yellow','green','green'])
```
